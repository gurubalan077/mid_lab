# Morphological Operations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("g.png",0)
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
```

```python
kernel = np.ones((3,3),np.uint8)
```

```python
erosion = cv2.erode(img , kernel , iterations = 2)
```

```python
dilation = cv2.dilate(img , kernel , iterations = 2)
```

```python
open = cv2.morphologyEx(img , cv2.MORPH_OPEN , kernel)
```

```python
close = cv2.morphologyEx(img , cv2.MORPH_CLOSE , kernel)
```

```python
gradient =  cv2.morphologyEx(img , cv2.MORPH_GRADIENT , kernel)
```

```python
tophat =  cv2.morphologyEx(img , cv2.MORPH_TOPHAT , kernel)
```

```python
blackhat =  cv2.morphologyEx(img , cv2.MORPH_BLACKHAT, kernel)
```

```python
tities = ["Original img" , "Erosion" , "Dilation" ," open" , " Close" , "Gradient" , "tophat" , "blackhat" ]
image = [ img , erosion , dilation , open , close , gradient , tophat , blackhat]

for i in range(len(image)):
    plt.subplot(2 , 4 ,i +1)
    plt.imshow(image[i])
    plt.title(tities[i])
    plt.xticks([]) , plt.yticks([])

plt.show()
    
```

# Image segmentation using Watershed Algorithm

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread("coin.jpg")
```

```python
plt.imshow(image)
plt.show()
```

```python
img_RGB = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
plt.show()
```

```python
image1 = cv2.medianBlur(img_gray,3)
ath2 = cv2.adaptiveThreshold(image1 , 255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,11)
plt.imshow(ath2)
plt.show()
```

```python
kernal = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(ath2,cv2.MORPH_OPEN,kernal , iterations = 13)
plt.imshow(opening)
plt.show()
```

```python
sure_bg = cv2.dilate(opening,kernal,iterations = 7)
plt.imshow(sure_bg)
plt.show()
```

```python
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2 ,3)
plt.subplot(1,2,1),plt.imshow(dist_transform)
plt.title("Dist_transform")
ret , sure_fg = cv2.threshold(dist_transform,0.05*dist_transform.max(),255 , 0)
plt.subplot(1,2,2),plt.imshow(sure_fg)
plt.title('Sobel X ')
plt.show()
```

```python
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)
plt.imshow(unknown)
plt.show()
```

```python
ret , markers = cv2.connectedComponents(sure_bg)
markers = markers+1
plt.imshow(markers)
plt.show()
```

```python
markers[unknown==255] = 0
plt.imshow(markers)
plt.show()
```

```python
markers = cv2.watershed(image,markers)
image[markers == -1] = (255 , 0 , 0)
plt.imshow(image)
plt.axis('off')
plt.show()
```

# Image segmentation using Mean shift algorithm

```python
image = cv2.imread("image_path.jpg")
```

```python
spatial_radius = 10
color_radius = 20
max_pyramid_level = 2
```

```python
shifted_image = cv2.pyrMeanShiftFiltering(image, sp=spatial_radius, 
sr=color_radius, maxLevel=max_pyramid_level)
```

```python
cv2.imshow("Original Image", image)
cv2.imshow("Mean Shift Segmented Image", shifted_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
